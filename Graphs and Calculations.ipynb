### Options Greeks

```python
import numpy as np
from scipy.stats import norm
```

```python
#  Functions that return d_1, d_2 and call and put prices
def d(sigma, S, K, r, t):
    d1 = 1 / (sigma * np.sqrt(t)) * ( np.log(S/K) + (r + sigma**2/2) * t)
    d2 = d1 - sigma * np.sqrt(t)
    return d1, d2

def call_price(sigma, S, K, r, t, d1, d2):
    C = norm.cdf(d1) * S - norm.cdf(d2) * K * np.exp(-r * t)
    return C

def put_price(sigma, S, K, r, t, d1, d2):
    P = -norm.cdf(-d1) * S + norm.cdf(-d2) * K * np.exp(-r * t)
    return P

#  Functions for Deltam Gamma, and  Theta
def delta(d_1, contract_type):
    if contract_type == 'c':
        return norm.cdf(d_1)
    if contract_type == 'p':
        return -norm.cdf(-d_1)
    
def gamma(d2, S, K, sigma, r, t):
    return( K * np.exp(-r * t) * (norm.pdf(d2) / (S**2 * sigma * np.sqrt(t) ))) 

def theta(d1, d2, S, K, sigma, r, t, contract_type):
    if contract_type == 'c':
        theta = -S * sigma * norm.pdf(d1) / (2 * np.sqrt(t)) - r * K * np.exp(-r * t) * norm.cdf(d2)
    if contract_type == 'p':
        theta = -S * sigma * norm.pdf(-d1) / (2 * np.sqrt(t)) + r * K * np.exp(-r * t) * norm.cdf(-d2)

    return theta

def vega(sigma, S, K, r, t):
    d1, d2 = d(sigma, S, K, r, t)
    v = S * norm.pdf(d1) * np.sqrt(t)
    return v
```

```python
greeks_puts = pd.DataFrame()
greeks_puts['strike'] = spy_puts['strike']
r = 0.01
t = 60/365

greeks_puts['d1'] = d(spy_puts['impliedVolatility'], spy_price, spy_puts['strike'], 0.01, 60/365)[0]
greeks_puts['d2'] = d(spy_puts['impliedVolatility'], spy_price, spy_puts['strike'], 0.01, 60/365)[1]
greeks_puts['delta'] = delta(greeks_puts['d1'], 'p')
greeks_puts['gamma'] = gamma(greeks_puts['d2'], spy_price, spy_puts['strike'], spy_puts['impliedVolatility'], r, t)
greeks_puts['openInterest'] = spy_puts['openInterest']
greeks_puts['Options Gravity'] = greeks_puts['openInterest'] * greeks_puts['gamma']
greeks_puts.fillna(0, inplace=True)

display(greeks_puts)
```

```python
greeks_calls = pd.DataFrame()
greeks_calls['strike'] = spy_calls['strike']
r = 0.01
t = 60/365

greeks_calls['d1'] = d(spy_calls['impliedVolatility'], spy_price, spy_calls['strike'], 0.01, 60/365)[0]
greeks_calls['d2'] = d(spy_calls['impliedVolatility'], spy_price, spy_calls['strike'], 0.01, 60/365)[1]
greeks_calls['delta'] = delta(greeks_calls['d1'], 'c')
greeks_calls['gamma'] = gamma(greeks_calls['d2'], spy_price, spy_calls['strike'], spy_calls['impliedVolatility'], r, t)
greeks_calls['openInterest'] = spy_calls['openInterest']
greeks_calls['Options Gravity'] = greeks_calls['openInterest'] * greeks_calls['gamma']
greeks_calls.fillna(0, inplace=True)

display(greeks_calls)
```

```python
plt.rcParams['figure.figsize'] = (20,10)
fig, ax = plt.subplots()
ax.title.set_text('Total SPY Options Gravity - June 17 2022 EXPIRY')
ax.bar(spy_calls['strike'], greeks_calls['Options Gravity'], label='Calls Gravity', color='green')
ax.bar(spy_puts['strike'], greeks_puts['Options Gravity'], label='Put Gravity', color='red')
ax.vlines(x=spy_price, ymin=0, ymax=max(greeks_calls['Options Gravity'].max(), greeks_puts['Options Gravity'].max()), label='Current Price', colors='blue')
ax.set_xlabel('Strike Price')
ax.set_ylabel('Options Gravity')
plt.legend(loc='upper left')
sns.despine()
plt.show()
```
